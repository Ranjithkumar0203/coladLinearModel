```
import torch
import torch.nn as nn

class Model(nn.Module):
    def __init__(self, inputCollumsInMatric):
        super(Model, self).__init__()
        self.linear1 = nn.Linear(inputCollumsInMatric, 3)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(3, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, inputCollumsInMatric):
      out = self.linear1(inputCollumsInMatric)
      out = self.relu(out)
      out = self.linear2(out)
      out = self.sigmoid(out)
      return out
```

```
features = torch.rand(10, 5)
```

```
model = Model(features.shape[1])
```

```
model(features)
```
